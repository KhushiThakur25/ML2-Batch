```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting a dashed line
plt.plot(x, y, linestyle='--', color='blue')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Dashed Line Plot')
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting a dashed line
plt.plot(x, y, linestyle=':', color='blue')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Dot Line Plot')
plt.grid(True)
plt.show()
```

```python
names = ['John','Shan','Max','Smith','Nick','Steve']
data = {
    "names" : names,
    "dept" : ["IT","HR","IT","HR","Admin","IT"],
    "salary" : [56000,50000,45000,32000,70000,30000]}
```

```python
plt.figure(figsize=(10,6))
c = ["y","b","m","g","c","r"]
plt.barh(data['names'], data['salary'],align = 'center',color = "red",edgecolor = "c",linewidth = 2,label = "Salary Growth")
plt.title('Salary of Employees')
plt.xlabel('Names',fontsize = 15)
plt.ylabel('Salary',fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.pie(data['salary'], labels = data['names'], autopct = "%1.2f%%", startangle = 90,counterclock = False,explode=(0,0.4,0,0,0,0),shadow = True,colors = ["green","blue","pink","purple","red","yellow"])
plt.title("Salary of Employees")
plt.legend()
plt.show()
```
